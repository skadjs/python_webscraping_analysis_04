```python
# 0. 기본 세팅
# 필요 라이브러리 import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 0. 기본 세팅
# 폰트 및 한글 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 0. 기본 세팅
# 분석할 csv 파일 불러오기
file_path = "/Users/eon/skRookies/myPython/Python_WebScraping_Analysis/data/인구현황.csv"
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
# 1. 지역별 총인구수 Top5
# 필터링 - 결괏값에 "전국" 미포함
df = df[df['행정기관'] != '전국']
```

```python
# 1. 지역별 총인구수 Top5
# top5 뽑기(전국 제외)
top5_pop = (
    df[['행정기관', '총인구수']]
    .sort_values('총인구수', ascending=False)
    .head(5) \
)
```

```python
# 1. 지역별 총인구수 Top5
# 결과 출력
print(top5_pop)
```

```python
# 1-1. 지역별 총인구수 plot
# 지역별 총인구수 뽑기(전국 제외)
regions_population = (
    df[['행정기관', '총인구수']]
    .sort_values('총인구수', ascending=False)
)
```

```python
# 1-1. 지역별 총인구수 plot
# 시각 자료 생성
plt.figure(figsize=(10, 5))
sns.barplot(
    data=regions_population,
    x='행정기관',
    y='총인구수',
    palette='pastel')
plt.title('지역별 총인구수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
