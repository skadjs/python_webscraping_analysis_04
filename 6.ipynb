```python
# 0. 기본 세팅
# 필요 라이브러리 import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 0. 기본 세팅
# 폰트 및 한글 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 0. 기본 세팅
# 분석할 csv 파일 불러오기
file_path = "/Users/eon/skRookies/myPython/Python_WebScraping_Analysis/data/인구현황.csv"
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
# 0. 기본 세팅
# 필터링 - 결괏값에 "전국" 미포함
df = df[df['행정기관'] != '전국']
```

```python
# 6. 남자 인구 vs 여자 인구 비교 Plot
transformed_df = df.melt(
    id_vars='행정기관',
    value_vars=['남자 인구수', '여자 인구수'],
    var_name='',
    value_name='인구수'
)
```

```python
# 6. 남자 인구 vs 여자 인구 비교 Plot
# 시각 자료 생성
plt.figure(figsize=(16,6))
sns.barplot(
    data=transformed_df,
    x='행정기관',
    y='인구수',
    hue=''
)
plt.xticks(rotation=45)
plt.title('남자 vs 여자 인구 비교')
```
