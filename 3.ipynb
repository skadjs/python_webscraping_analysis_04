```python
# 0. 기본 세팅
# 필요 라이브러리 import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 0. 기본 세팅
# 폰트 및 한글 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 0. 기본 세팅
# 분석할 csv 파일 불러오기
file_path = "/Users/eon/skRookies/myPython/Python_WebScraping_Analysis/data/인구현황.csv"
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
# 0. 기본 세팅
# 필터링 - 결괏값에 "전국" 미포함
df = df[df['행정기관'] != '전국']
```

```python
# 3. 남녀 비율 분석
# 지역별 남녀비율 테이블 뽑아내기
gender_ratio = (
    df[['행정기관', '남여 비율']]
    .sort_values('남여 비율', ascending=False)
    .reset_index(drop=True)
)
```

```python
# 3. 남녀 비율 분석
# 결과 출력
print(gender_ratio)
```

```python
# 3-1. 남녀 비율 분석 Plot
# 시각 자료 생성
plt.figure(figsize=(10, 5))
sns.barplot(
    data=gender_ratio,
    x='행정기관',
    y='남여 비율',
    palette='Set2')
plt.title('지역별 남녀 비율')
plt.xticks(rotation=45)
plt.axhline(
    y = 1.0,
    color = 'red',
    linestyle = '--',
    label = '남녀 1:1 비율'
)
plt.legend()
plt.tight_layout()
plt.show()
```
