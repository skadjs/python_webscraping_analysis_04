```python
# 0. 기본 세팅
# 필요 라이브러리 import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 0. 기본 세팅
# 폰트 및 한글 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 0. 기본 세팅
# 분석할 csv 파일 불러오기
file_path = "/Users/eon/skRookies/myPython/Python_WebScraping_Analysis/data/인구현황.csv"
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
# 0. 기본 세팅
# 필터링 - 결괏값에 "전국" 미포함
df = df[df['행정기관'] != '전국']
```

```python
# 2. 세대당 인구수 높은 지역 Top5
# Top5 뽑기
top5_per_house = (
    df[['행정기관', '세대당 인구']]
    .sort_values('세대당 인구', ascending=False)
    .head(5)
    .reset_index(drop=True)
)
```

```python
# 2. 세대당 인구수 높은 지역 Top5
# 결과 출력
print(top5_per_house)
```

```python
# 2-1. 지역별 세대당 인구 Plot
# 지역별 세대당 인구 뽑아내기
per_house = (
    df[['행정기관', '세대당 인구']]
    .sort_values('세대당 인구', ascending=False)
)
```

```python
# 2-1. 지역별 세대당 인구 Plot
# 시각 자료 생성
plt.figure(figsize=(10, 5))
sns.barplot(
    data=per_house,
    x='행정기관',
    y='세대당 인구',
    palette='viridis'
    )
plt.title('지역별 세대당 인구')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
